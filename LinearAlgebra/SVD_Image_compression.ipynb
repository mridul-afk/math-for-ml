```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("image.jpg")
X = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
U, S, V_T = np.linalg.svd(X, full_matrices=False)
S = np.diag(S)
```

```python
ranks = [5, 20, 50, 100]
plt.imshow(X, cmap='gray')
plt.title("Original")
plt.axis('off')
```

```python
plt.figure()

plt.subplot(1, len(ranks) + 1, 1)
plt.imshow(X, cmap='gray')
plt.title("Original")
plt.axis('off')

for i, r in enumerate(ranks):
    X_approx = U[:, :r] @ S[:r, :r] @ V_T[:r, :]

    plt.subplot(1, len(ranks) + 1, i + 2)
    plt.imshow(X_approx, cmap='gray')
    plt.title(f"r = {r}")
    plt.axis('off')

plt.show()
```
